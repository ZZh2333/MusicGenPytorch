```python
import requests
import re
from bs4 import BeautifulSoup as bs
```

```python
url = "http://www.51windows.net/share/midi/cmidi.htm"
```

```python
list = []
download_preurl = "http://www.51windows.net/share/midi/"
r= requests.get(url)
rt = r.content
rh = str(rt,"GB2312","ignore")
soup = bs(rh,"html.parser")
download = soup.find_all("a")
for i in download:
    if(i['href'][:3]=="i8_"):
        # print(i['href'])
        list.append(download_preurl+i['href'])
list
```

```python
for i,u in enumerate(list):
    print(list[i].split("/")[6])
    data = requests.get(list[i])
    with open("dataset/"+str(list[i].split("/")[6]),"wb") as code:
        code.write(data.content)
```
