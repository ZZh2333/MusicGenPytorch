```python
import requests
import re
from bs4 import BeautifulSoup as bs
```

```python
url_pre = "https://www.midiworld.com/search/"
```

```python
list = []
for i in range(1,40):
    url = url_pre + str(i) + "/?q=pop"
    r = requests.get(url)
    rt = r.content
    rh = str(rt,"utf-8")
    soup = bs(rh,"html.parser")
    download = soup.find_all("a",attrs={"target":"_blank"})
    for j in download:
        list.append(j['href'])
list
```

```python
len(list)
```

```python
for i,u in enumerate(list):
    data = requests.get(list[i])
    with open("midi_dataset/demo"+str(i)+".mid","wb") as code:
        code.write(data.content)
```
